```r
# Simulando o jogo com dado honesto
set.seed(1)
N = 1000 # 1000 repeticoes
vitA = 0 # iniciando objeto que guardara o numero de vitorias de A em N simulacoes
vitB = 0 # iniciando objeto que guardara o numero de vitorias de B em N simulacoes
for(i in 1:N)
{
  DadoA = sample(1:6,1)
  if(DadoA==6)
  {
  vitA = vitA +1
  }else{
  DadoB = sample(1:6,1)  
  if(any(DadoB==c(5,6)))
  {
  vitB = vitB+1
  }else{
    DadoA = sample(1:6,1)
    if(any(DadoA==c(4,5,6)))
    {
      vitA = vitA +1
    }else{
      DadoB = sample(1:6,1)  
      if(any(DadoB==c(3,4,5,6)))
      {
        vitB = vitB+1
      }else{
        DadoA = sample(1:6,1)
        if(any(DadoA==c(2,3,4,5,6)))
        {
          vitA = vitA +1
        }else{
          DadoB = sample(1:6,1)  
          if(any(DadoB==c(1,2,3,4,5,6)))
          {
            vitB = vitB+1
            }
}
      }
    }
  }
  }
}

vitA/N
vitB/N
```

```r
# ------------------------- #
# Problema dos aniversarios #
# ------------------------- #

# Solucao exata

birthday=function(x){
  a=1-exp(-(x^2)/(2*365))
  return(a)
}
birthday(23)

plot(1:150,birthday(1:150),bty="n",pch=19,cex=0.4,xlab="n",ylab="Probabilidade")
abline(v=23,col=2,lty=2)
abline(h=0.5,col=4,lty=2)

# Simulacao

N = 1000                                   # Numero de simulacoes do experimento

prob = c()
for(n in 2:100){                            # Para n variando de 2 atĂŠ 100
  cont_a = 0                                # Inicia o contador
  M = matrix(NA, N, n)                      # Delara uma matriz varia com as dimensoes desejadas  
  for(i in 1:N){                            # indice i que percorre todas as N linhas simuladas
    M[i,] = sample(1:365, n, replace = T)   # Sorteio de uma amostra de tamanho n de numeros de 1 a 365 
    tab = table(M[i,])                      # objeto tab guarda a tabela de frequencias dessa amostra
    if(any(tab>1)){                         # se algum elemento da tabela de frequencias for maior que 1
      cont_a = cont_a+1                     # contador recebe 1 pois duas pessoas fizeram aniversario no mesmo dia
    } 
  }
  prob[n] = cont_a/N                        # a probabilidade serĂĄ a proporcao de pessoas que fazem aniversario no mesmo dia observadas em N amostra simuladas
}

prob

prob[23]

# Comparando ambos
plot(1:100,birthday(1:100),bty="n",pch=19,cex=0.4,xlab="n",ylab="Probabilidade")
lines(1:100,prob,col=2)
```
